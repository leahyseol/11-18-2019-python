```python
from collections import Counter

f=open("yesterday.txt",'r')
txt_yesterday=f.readlines()
#print(txt_yesterday)

contents=""
for line in txt_yesterday:
    contents+=line.strip()+"\n"
#print(contents)
yesterday_list=contents.split()
#print(yesterday_list)
c=Counter(yesterday_list)
#print(c.items())
for k,v in c.items():
    print(k,":",v)
```
