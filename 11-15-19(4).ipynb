```python
import numpy as np
```

```python
data=np.loadtxt("data/ratings1.csv",delimiter=",",dtype=np.float64)
print(data[:5,:])
print(data.shape)
```

```python
mean_rating=data[:,2].mean()
print(mean_rating)
```

```python
movie_ids=np.unique(data[:,1])
print(movie_ids)
print(len(movie_ids))
```

```python
user_ids=np.unique(data[:,0])
```

```python
mean_values=[]
for user_id in user_ids:
    data_for_user=data[data[:,0]==user_id,:]
    value=data_for_user[:,2].mean()
    mean_values.append([user_id,value])
mean_values[:5]
```

```python
mean_values=[] #for(String :list)
for movie_id in movie_ids:
    data_for_movie=data[data[:,1]==movie_id,:]
    value=data_for_movie[:,2].mean()
    mean_values.append([movie_id,value])
mean_values[:5]
```
